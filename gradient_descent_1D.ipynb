## Gradient Descent 1D


```python
# Imports
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_inline import backend_inline

backend_inline.set_matplotlib_formats("svg")

import sympy as sym
import sympy.plotting.plot as symplot
```

```python
# Function
def fx(x):
    return 3 * x**2 - 3 * x + 4


# Derivative
def dfx(x):
    return 6 * x - 3
```

```python
# Plot function and it's derivative
# define a range for x
x = np.linspace(-2, 2, 2001)

# Plotting
plt.plot(x, fx(x), x, dfx(x))
plt.xlim(x[[0, -1]])
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["y", "dy"])
plt.show()
```

```python
# Random starting point
local_min = np.random.choice(x, 1)
print("Starting point of local_min: ", local_min)
# Learning Parameters
lr = 0.01
traning_epochs = 100

# Run through training
for i in range(traning_epochs):
    grad = dfx(local_min)
    local_min = local_min - lr * grad

print("local_min after gradient descent: ", local_min)
```

```python
# Plot the results
plt.plot(x, fx(x), x, dfx(x))
plt.plot(local_min, dfx(local_min), "ro")
plt.plot(local_min, fx(local_min), "ro")
plt.xlim(x[[0, -1]])
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["f(x)", "df", "f(x) min"])
plt.title("Empirical local minimum: %s " % local_min[0])
plt.show()
```

```python
# Store the model parameters and outputs on each iteration
# Random starting point
local_min = np.random.choice(x, 1)
print("Starting point of local_min: ", local_min)
# Learning Parameters
lr = 0.01
traning_epochs = 100

# Run through training and store all the results
model_params = np.zeros((traning_epochs, 2), dtype=object)
for i in range(traning_epochs):
    grad = dfx(local_min)
    local_min = local_min - lr * grad
    model_params[i, :] = local_min, grad
```

```python
# Plot the gradient over iterations
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

for i in range(2):
    ax[i].plot(model_params[:, i], "o-")
    ax[i].set_xlabel("Iteration")
    ax[i].set_title(f"Final estimated minimum: {local_min[0]:0.5f}")

ax[0].set_ylabel("Local minimum")
ax[1].set_ylabel("Derivative")

plt.show()
```

## Find minimum of following function

    f(x) = cos(2*pi*x) + x**2
    Hard code starting value x = 0


```python
x = sym.symbols("x")


def fx(x):
    return np.cos(2 * np.pi * x) + x**2


def dfx(x):
    return -(np.sin(2 * np.pi * x) * (2 * np.pi) + 2 * x)
```

```python
p = symplot(fx, (x, -4, 4), label="Function", show=False, line_color="blue")
p.extend(symplot(dfx, (x, -4, 4), label="Derivative", show=False, line_color="red"))
p.legend = True
p.title = "The Function and Derivatives"
p.show()
```

```python
x = np.linspace(-2, 2, 2001)

# Plotting
plt.plot(x, fx(x), x, dfx(x))
plt.xlim(x[[0, -1]])
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["y", "dy"])
plt.show()
```

```python
# Store the model parameters and outputs on each iteration
# Random starting point
local_min = np.random.choice(x, 1)
print("Starting point of local_min: ", local_min)
# Learning Parameters
lr = 0.01
traning_epochs = 100

# Run through training and store all the results
model_params = np.zeros((traning_epochs, 2), dtype=object)
for i in range(traning_epochs):
    grad = dfx(local_min)
    local_min = local_min - lr * grad
    model_params[i, :] = local_min, grad

# Plot the gradient over iterations
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

for i in range(2):
    ax[i].plot(model_params[:, i], "o-")
    ax[i].set_xlabel("Iteration")
    ax[i].set_title(f"Final estimated minimum: {local_min[0]:0.5f}")

ax[0].set_ylabel("Local minimum")
ax[1].set_ylabel("Derivative")
```
