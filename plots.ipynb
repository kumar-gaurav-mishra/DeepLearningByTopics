# Dots, lines, plots, subplots, images etc...

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Point plot
plt.plot(10, 10, 'bo', label='blue dot')
plt.plot(4, 4, 'r^', label="red triangle")
plt.plot(2, 9, 'gp', label='green pentagon')
plt.plot(7, 3, 'ys', label='yellow square') 
# plt.legend() # we can provide label in this way or directly as an array in legend function but order need to be maintained

plt.legend(['blue', 'red', 'green', 'yello']) # This over rides the label field in the plots
plt.show()
```

```python
# Line Plot
x = [0, 2]
y = [0, 2]
plt.plot(x, y, 'r')
```

```python
x = [0, 2, 4]
y = [0, 2, -2]
plt.plot(x, y, 'r')
plt.show()
plt.plot(x, y, 'rs')
plt.show()
plt.plot(x, y, 'rs-')
plt.show()
plt.plot(x, y, 'rs--')
plt.show()
plt.plot(x, y, 'rs:')
plt.show()
```

```python
x = np.linspace(0, 3*np.pi, 1001)
y = np.sin(x)

plt.plot(x, y, 'm:')
plt.legend(['Sin line'])
plt.show()
```

```python
# Draw X using plot
x1 = [0, 1,]
y1 = [0, 1]
x2 = [0, 1]
y2 = [1, 0]
x3 = [0, 1.01]
y3 = [1.01, 0]
plt.plot(x1, y1, 'r', linewidth=15)
plt.legend('X')
plt.plot(x2, y2, 'g')
plt.plot(x3, y3, 'b')
plt.axis('square')
plt.show()
```

```python
# Subplots
plt.subplot(3,3,1)
plt.plot(np.random.rand(20), 'm')
plt.subplot(3,3,2)
plt.plot(np.random.rand(20), 'b')
plt.subplot(3,3,3)
plt.plot(np.random.rand(20), 'g')
plt.subplot(3,3,4)
plt.plot(np.random.rand(20), 'r')
plt.subplot(3,3,5)
plt.plot(np.random.rand(20), 's--')
plt.subplot(3,3,6)
plt.plot(np.random.rand(20), 'r:')
```
